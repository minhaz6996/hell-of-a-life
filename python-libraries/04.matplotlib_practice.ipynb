# Subplot

```python
i=np.array([1,3,2,5,4])
j=np.array([2,5,3,7,1])
c=np.array([3,5,8,2,4])
d=np.array([4,7,5,9,1])
```

```python
plt.subplot(2,1,1)
```

```python
plt.subplot(1,4,2)
```

```python
plt.plot(i,j)
plt.subplot(1,2,2)
plt.plot(c,d)
plt.subplot(1,2,2)

plt.show()
```

```python
plt.plot(i,j)
plt.subplot(1,2,2)
plt.plot(c,d)

plt.show()
```

```python
plt.plot(i,j)
plt.subplot(2,1,2)
plt.plot(c,d)
plt.subplot(2,1,2)

plt.show()
```

# Add Title to Plot

```python
plt.plot(i,j)
plt.subplot(1,2,2)
plt.title('First plot')
plt.plot(c,d)
plt.subplot(1,2,2)
plt.title('Second plot')
plt.suptitle('Viewing Two plots at a time')

plt.show()
```

# Scatter Plot

```python
m=np.array([3,4,5,6,7,8,9])
n=np.array([1,4,5,7,8,9,12])
plt.scatter(m,n)
```

```python
plt.grid()
```

```python
plt.plot(m,n)
plt.xlabel('value for m')
plt.ylabel('value for n')
plt.title('Random Scatterplot')
plt.grid()
```

```python
plt.show()
```

```python
colors=np.array(['pink','yellow','green','brown','blue','black','red'])
plt.grid()
plt.scatter(m,n,c=colors)
```

```python
size=np.array([12,16,23,80,19,30,32])

plt.scatter(m,n,s=size)
```
