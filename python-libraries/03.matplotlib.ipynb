```python
a=np.array([2,3,4,5,1,6])
b=np.array([3,4,5,2,1,7])
c=np.array([6,8,3,7,4,2])
d=np.array([2,3,4,5,1,6])

plt.subplot(1,2,1)
```

```python
plt.plot(a,b)
plt.subplot(1,2,2)
plt.plot(c,d)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(a,b)
plt.subplot(2,1,2)
plt.plot(c,d)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(a,b)
plt.title('First Subplot')
plt.subplot(1,2,2)
plt.plot(c,d)
plt.plot('Second Subplot')
plt.suptitle('Two Plots')
plt.show()
```

# Multivariate Analysis = Scatterplot

```python
x=np.array([4,5,6,3,8,3])
y=np.array([5,6,9,2,5,3])

plt.scatter(x,y)
```

```python
colors=np.array(['green','red','yellow','blue','pink','brown'])
plt.scatter(x,y, c=colors)
```

```python
size=np.array([8,12,80,40,23,98])
plt.scatter(x,y,s=size)
```
